```python
df = pd.read_csv("../content/hr_data_preprocessing.csv", encoding="utf-8", index_col=0)
```

## 1. Separate train and test data

```python
features = ["job_satisfaction",
            "working_hours",
            "num_companies_worked",
            "age",
            "years_at_company",
            "salary_increasing_rate",
            "department_hr",
            "department_rnd",
            "department_sales",
            "marital_status_etc",
            "marital_status_married",
            "marital_status_single"
            ]
```

```python
x_data = df[features]
y_data = df["attrition"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data,
                                                    test_size=0.25,
                                                    random_state=42,
                                                    stratify=y_data)
```

```python
print("test데이터의 비율:", len(y_test)/len(df))
```

```python
x_train.index
```

```python
train_indx = x_train.index
test_indx = x_test.index
```

```python
print("전체 데이터의 퇴직자 비율:", y_data.sum()/len(df))
print("test 데이터의 퇴직자 수:", y_test.sum())
print("test 데이터의 퇴직자 비율:", y_test.sum()/len(y_test))
```

##2. Data scaling

```python
#
minmax_scaler = MinMaxScaler()
x_train = minmax_scaler.fit_transform(x_train)
```

```python
x_train_scaled = pd.DataFrame(x_train, columns=features, index=train_indx)
```

```python
x_train_scaled
```

```python
x_test = minmax_scaler.transform(x_test)
```

```python
x_test_scaled = pd.DataFrame(x_test, columns=features, index=test_indx)
```

```python
train = x_train_scaled.merge(y_train, left_index=True, right_index=True)
test = x_test_scaled.merge(y_test, left_index=True, right_index=True)
```

```python
train.to_csv("../content/hr_train_scaling.csv", encoding="utf-8")
test.to_csv("../content/hr_test_scaling.csv", encoding="utf-8")
```
