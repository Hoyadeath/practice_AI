california_housing 데이터셋 : 1990년 캘리포니아 인구 조사 데이터를 기반으로 하며, 주택 가격을 예측하기 위한 특성들을 포함한 데이터

MedInc: 블록 그룹의 중간 소득 (수만 달러 단위)

HouseAge: 블록 그룹의 중간 주택 연령

AveRooms: 블록 내 주택 1개당 평균 방 개수

AveBedrms: 블록 내 주택 1개당 평균 침실 개수

Population: 블록 그룹 내 인구

AveOccup: 블록 내 주택 1개당 평균 거주자 수

Latitude: 블록 그룹의 위도

Longitude: 블록 그룹의 경도


**MedHouseVal: 목표 변수(Target Variable)**
  중간 주택 가치(중간 주택 가격). 단위 10만 달러

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fetching the California housing data
california_housing = fetch_california_housing()
california_df = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
california_df['MedHouseVal'] = california_housing.target
```

```python
print(california_df.head())
print(california_df.describe())
print(california_df.isnull().sum())
```

```python
california_df.hist(edgecolor='black', linewidth=0.1)
plt.show()
```

```python
california_df.boxplot()
plt.show()
```

```python
sns.pairplot(california_df, hue='Population')
plt.show()
```

```python
sns.heatmap(california_df.corr(), annot=True, cmap='cividis')
plt.show()
```

```python
sns.countplot(x='Latitude', data=california_df)
plt.show()
```

```python
sns.violinplot(x='Population', y='Latitude', data=california_df)
plt.show()
```

```python
sns.scatterplot(x='Population', y='Latitude', hue='Longitude', data=california_df)
plt.show()
```
