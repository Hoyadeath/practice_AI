아이리스(Iris) 데이터셋의 첫 세 가지 변수, 꽃받침의 길이(sepal length), <br> 꽃받침의 너비(sepal width), 그리고 꽃잎의 길이(petal length)를 사용 <br>
꽃잎의 너비(petal width)를 예측하는 다중 선형 회귀 모델 개발

```python
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Load the iris dataset
iris = datasets.load_iris()
X = iris.data[:, :3]  # we only take the first three features.
y = iris.data[:, 3]   # target is the petal width
```

```python
# Split the data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Create linear regression object
regr = LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
```

```python
# The coefficients
print('Coefficients:', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f' % mean_squared_error(y_test, y_pred))
# The coefficient of determination
print('Coefficient of determination: %.2f' % r2_score(y_test, y_pred))
```

```python
# Plotting the results
# This is a multi-dimensional space, so we can't plot it directly. We'll plot the true vs predicted values instead.
plt.scatter(y_test, y_pred, color='blue')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True vs Predicted Values for Petal Width')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', linewidth=2) # Diagonal line
plt.show()
```
