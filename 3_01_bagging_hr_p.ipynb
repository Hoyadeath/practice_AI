```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
# 데이터 불러오기
data = pd.read_csv("/content/hr_data.csv")


# 데이터 확인
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# 타겟 변수 인코딩
le = LabelEncoder()
data['attrition'] = le.fit_transform(data['attrition'])

```

```python
# 수치형 변수만 선택 ... 간단한 진행을 위해
numeric_features = data.select_dtypes(include=['int64', 'float64']).columns
X = data[numeric_features].dropna(axis=1)  # 누락된 값이 있는 열 제거
y = data['attrition'] # 목표 변수. 퇴직여부

# 훈련 세트와 테스트 세트로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# 결정 트리 분류기를 기본 모델로 사용하는 배깅 분류기 생성
bagging_clf = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=100,
    random_state=42)
```

```python
# 모델 학습
bagging_clf.fit(X_train, y_train)
# 테스트 세트에서의 성능 평가
y_pred = bagging_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'  ....... JYH.......   Test accuracy: {accuracy:.4f}')
```
