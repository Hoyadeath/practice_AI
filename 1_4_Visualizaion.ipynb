### Linear Charts by Series

```python
dataset=[[25, 100], [52, 256], [38, 152], [32, 140], [25, 150], [45, 183], [40, 175], [55, 203],
[28, 152], [42, 198]]
actual=[j[1] for j in dataset]
predicted=[115, 242, 176, 148, 115, 209, 185, 256, 129, 195]
error=[(x - y) for x, y in zip(actual, predicted)]
xs=[i for i, _ in enumerate(error)]
plt.plot(xs, actual, 'g-', label='x') # a solid green line
plt.plot(xs, predicted, 'b-', label='y') # a solid blue line
plt.plot(xs, error, 'r--', label='error') # a dotted red line
plt.legend(loc=9) # legend display:loc number positioning
plt.xlabel('mode')
plt.title('actual vs predicted')
plt.show()
```

##2. Bar charts, histograms, scatterplots, circular charts
### Bar chart

```python
from matplotlib import pyplot as plt
movies=['cinema', 'avengers', 'squid game', '007', 'iron man']
score=[5, 10, 3, 8, 9]
plt.bar(movies, score)
plt.ylabel('score')
plt.title('Movies')
plt.xticks([i+0.1 for i, _ in enumerate(movies)], movies) # Start drawing bar charts at 0.1
plt.show()
```

```python
plt.barh(movies,score)
plt.ylabel('Movies')
plt.title('score')
plt.yticks([i + 0.1 for i, _ in enumerate(movies)], movies) # Start drawing y-axis from 0.1
plt.show()
```

```python
plt.bar(movies, score, color='red', width=0.3)
plt.ylabel('score')
plt.title('Movies')
plt.xticks([i + 0.1 for i, _ in enumerate(movies)], movies)
plt.show()
```

### Histogram

```python
from collections import Counter
grades=[83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
histogram=Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar([x+5 for x in histogram.keys()], # Move 5 to the right and start drawing
histogram.values(),
10,
edgecolor=(0, 0, 0))
plt.axis([-5, 105, 0, 5]) # Specify x-axis and y-axis ranges
plt.xticks([10 * i for i in range(11)])# x-axis label display
plt.xlabel('Decile')
plt.ylabel('# of Students')
plt.title('Distribution of Exam 1 Grades')
plt.savefig('viz_grades.png') # Save Picture File
```

```python
from matplotlib import pyplot as plt
height=[175, 170, 162, 171, 186, 175, 166, 169, 189, 178] # x independent variable
weight=[70, 65, 61, 63, 78, 64, 60, 64, 88, 76] # y dependent variable
labels=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'I', 'j']
plt.scatter(height, weight)
for i, j, k in zip(labels, height, weight):
  plt.annotate(i,
               xy=(j, k), # Labeling
               xytext=(5, -5), # Text Location
  textcoords='offset points')
plt.title('height vs. weight')
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
import os
import glob
f1=[]
for file in glob.glob('*.txt'):
  with open(file, 'r', newline='') as filereader:
    for row in filereader:
      contents=row.split()
      for i in contents:
        contents=i.strip('.').strip(',').strip('')
        f1.append(contents)
f1
```

```python
import os
import glob

f1=[]
for file in glob.glob('*.txt'):
  with open(file, 'r', newline='') as filereader:
    for row in filereader:
      contents=row.split()
      for i in contents:
        contents=i.strip('.').strip(',').strip('').strip('를').strip('을').strip("''")
        contents=contents.strip('합니다').strip('하세요')
        contents=contents.strip('에')
        f1.append(contents)
f1
```

##2. Find the frequency of words


```python
from collections import Counter
f1
word_counts=Counter(f1)
dict(word_counts)
# word_counts.items()
wc=sorted(word_counts.items(),key=lambda count:count[1], reverse=True)
wc
```

```python
import pandas as pd
wcdf=pd.DataFrame(wc)
wcdf.columns=['word', 'freq']
```

```python
wcdf1=wcdf.set_index('word')
wcdf1.head()
```
