##3. Useful Data Frame Manipulation


```python
import pandas as pd
x=[0, 1, 2, 3, 4, 5]
y=[0, 2, 4, 6, 8, 10]
df=pd.DataFrame( )
df['x_value']=x
df['y_value']=y
df
```

```python
df['xy_value']=df['x_value'] * df['y_value']
df
```

###New Index: set_index function

```python
index=['zero', 'one', 'two', 'three', 'four', 'five']
df['number']=index
df
```

####[Keep existing indexes]

```python
df.set_index('number', append=True)
```

####[Clear existing indexes]

```python
df2=df.set_index('number', append=False)
df2
```

###Clear existing index and initialize: reset_index function

```python
df3=df2.reset_index(drop=True)
df3
```

###Create index by date: pd.date_range function


```python
dates=pd.date_range('20230101', periods=6)
dates
```

```python
df3=df2.reset_index(drop=True)
df4=df3.set_index(dates)
df4
```

###Extracting indexes with index attributes

```python
df4.index
```

###Extract row data with value attribute

```python
df4.values
```

###Charting: plot function

```python
df2.plot( )
```

###Sort by specific column: sort_values function

```python
df4.sort_values(by='x_value', ascending=False) # Sort in descending order by x_value column
```

###Add Mean Value column: mean function

```python
df3['mean']=df4.mean(axis=1)
df4
```
